# Scraping Contents

```python

from bs4 import BeautifulSoup
import requests
import string
df = pd.read_excel('Input.xlsx')
# Create a new column to store the extracted content
df['Extracted_Content'] = ""

# Loop through each URL
for index, row in df.iterrows():
    url = row['URL']  # Assuming the column name is 'URL'
    
    # Fetch the webpage content
    try:
        response = requests.get(url)
        soup = BeautifulSoup(response.content, 'html.parser')
        
        # Extract content
        content = soup.find(attrs={'class': 'td-post-content'}).text.replace('\n', ' ')
        
        # Remove punctuations
        content = content.translate(str.maketrans("", "", string.punctuation))
        print(content)
        # Update the DataFrame with the extracted content
        df.at[index, 'Extracted_Content'] = content
    except Exception as e:
        print(f"Error processing {url}: {str(e)}")

# End of loop


```

# Scrap stop words

```python
import nltk
from nltk.corpus import stopwords
stop_words_files = ["StopWords_Auditor.txt", "StopWords_Currencies.txt", "StopWords_DatesandNumbers.txt","StopWords_Generic.txt","StopWords_GenericLong.txt","StopWords_Geographic.txt","StopWords_Names.txt"]
all_stop_words = set()
for file in stop_words_files:
    with open(file, "r") as f:
        stop_words = f.read().splitlines()
        all_stop_words.update(stop_words)
```

# Tokenization and remove stop words

```python
import nltk
from nltk.tokenize import word_tokenize
df['Tokenized_Content'] = df['Extracted_Content'].apply(lambda x: word_tokenize(x))
stop_words = set(stopwords.words('english'))

# Remove stopwords from the tokenized content
df['Filtered_Content'] = df['Tokenized_Content'].apply(lambda tokens: [word for word in tokens if word.lower() not in stop_words])
```

# Positive Score

```python
with open("positive-words.txt", "r") as file:
    positive_words = set(file.read().splitlines())

# Create a function to filter positive words from a list of tokens
def filter_positive_words(Filtered_Content):
    return [word for word in Filtered_Content if word in positive_words]

# Apply the function to the 'Filtered_Content' column
df['Positive_Words'] = df['Filtered_Content'].apply(filter_positive_words)
df['POSITIVE_SCORE'] = df['Positive_Words'].apply(lambda x: len(str(x).split()))   
```

# Negative Score

```python
with open("negative-words.txt", "r") as file:
    negative_words = set(file.read().splitlines())

# Create a function to filter positive words from a list of tokens
def filter_negative_words(Filtered_Content):
    return [word for word in Filtered_Content if word in negative_words]

# Apply the function to the 'Filtered_Content' column
df['negative_Words'] = df['Filtered_Content'].apply(filter_negative_words)
df['NEGATIVE_SCORE'] = df['negative_Words'].apply(lambda x: len(str(x).split()))
```

# polarity_score

```python
def calculate_polarity_score(row):
    positive_score = (row['POSITIVE_SCORE'])
    negative_score = (row['NEGATIVE_SCORE'])
    polarity_score = (positive_score - negative_score) / ((positive_score + negative_score) + 0.000001)
    return polarity_score

# Apply the function to calculate polarity score
df['Polarity_Score'] = df.apply(calculate_polarity_score, axis=1)
```

# WORD COUNT

```python
df['WORD_COUNT'] = df['Filtered_Content'].apply(lambda x: len(str(x).split()))   
```

# subjectivity

```python
def calculate_subjectivity(row):
    positive_score = (row['POSITIVE_SCORE'])
    negative_score = (row['NEGATIVE_SCORE'])
    Total_Words_after_cleaning = (row['WORD_COUNT'])
    Subjectivity_Score = (positive_score + negative_score)/ ((Total_Words_after_cleaning) + 0.000001)
    return Subjectivity_Score

# Apply the function to calculate polarity score
df['SUBJECTIVITY SCORE'] = df.apply(calculate_subjectivity, axis=1)
```

# AVG SENTENCE LENGTH

```python
import re
def calculate_avg_sentence_length(row):
    content = row['Extracted_Content']
    sentences = re.split(r'[?!.]', content)
    total_words = len(content.replace(' ', ''))
    avg_sentence_length = total_words / len(sentences)
    return avg_sentence_length

# Apply the function to calculate average sentence length
df['AVG SENTENCE LENGTH'] = df.apply(calculate_avg_sentence_length, axis=1)
```

# PERCENTAGE OF COMPLEX WORDS

```python
def calculate_complex_word_percentage(row):
    content = row['Extracted_Content']
    total_words = len(content.split())
    if total_words == 0:
        return 0  # Avoid division by zero
    total_complex_words = sum(1 for word in content.split() if len(word) > 2)  
    complex_word_percentage = (total_complex_words / total_words) * 100
    return complex_word_percentage

# Apply the function to calculate complex word percentage
df['PERCENTAGE OF COMPLEX WORDS'] = df.apply(calculate_complex_word_percentage, axis=1)
```

# FOG INDEX

```python
def calculate_fog_index(row):
    Avg_Sentence_Length = row['AVG SENTENCE LENGTH']
    Complex_Word_Percentage = row['PERCENTAGE OF COMPLEX WORDS']

    fog_index = 0.4 * (Avg_Sentence_Length + Complex_Word_Percentage)
    return fog_index

# Apply the function to calculate FOG Index
df['FOG_Index'] = df.apply(calculate_fog_index, axis=1)
```

# AVG NUMBER OF WORDS PER SENTENCE

```python
def calculate_avg_words_per_sentence(row):
    content = row['Extracted_Content']
    sentences = re.split(r'[?!.]', content)
    total_words = len(content.split())
    total_sentences = len(sentences)
    avg_words_per_sentence = total_words / total_sentences
    return avg_words_per_sentence

# Apply the function to calculate average words per sentence
df['AVG NUMBER OF WORDS PER SENTENCE'] = df.apply(calculate_avg_words_per_sentence, axis=1)
```

# COMPLEX WORD COUNT

```python
def calculate_complex_word(row):
    content = row['Extracted_Content']
    total_words = len(content.split())
    total_complex_words = sum(1 for word in content.split() if len(word) > 2)  # Assuming words with more than 3 syllables are complex
    return total_complex_words

# Apply the function to calculate complex word percentage
df['COMPLEX WORD COUNT'] = df.apply(calculate_complex_word, axis=1)
```

# Word Count

```python
df['WORD COUNT'] = df['Filtered_Content'].apply(lambda x: len(str(x).split()))   
```

# SYLLABLE PER WORD

```python
def syllable_count(word):
    count = 0
    vowels = "AEIOUYaeiouy"
    if word[0] in vowels:
        count += 1
    for index in range(1, len(word)): 
        if word[index] in vowels and word[index - 1] not in vowels:
            count += 1
            if word.endswith("es") or word.endswith("ed"):
                count -= 1
    if count == 0:
        count += 1
    return count

# Calculate syllable count for each word in each row
for index, row in df.iterrows():
    syllable_counts = [syllable_count(word) for word in row['Filtered_Content']]
    df.at[index, 'SYLLABLE PER WORD'] = sum(syllable_counts)
```

# PERSONAL PRONOUNS

```python
import spacy
nlp = spacy.load('en_core_web_lg')
```

```python
df['PERSONAL PRONOUNS'] = np.nan

for i in range(len(df)):
    doc = nlp(df['Extracted_Content'][i])
    PRONOUNS = []
    for token in doc:
        if token.pos_ == 'PRON':
            PRONOUNS.append(token)
        
    df['PERSONAL PRONOUNS'][i] = PRONOUNS
```

# AVG WORD LENGTH

```python
def calculate_avg_words_Length(row):
    words = row['Tokenized_Content']
    if len(words) == 0:
        return 0
    total_characters = sum(len(word) for word in words)
    average_length = total_characters / len(words)
    return average_length

# Apply the function to calculate average words per sentence
df['AVG WORD LENGTH'] = df.apply(calculate_avg_words_Length, axis=1)
```
